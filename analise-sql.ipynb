# 1 - Introdução.

## 1.1 - Contexto.

A tabela foi criada no AWS Athena, junto com o S3 Bucket. Abaixo vamos mostrar as querys usadas para realizar a analise dos dados, junto com os resultados.

## 1.2 - Fonte

A fonte de dados usada foi disponibilizada no [link](https://github.com/andre-marcos-perez/ebac-course-utils/tree/main/dataset).

## 1.3 - Dados

Os dados representam informações de clientes de um banco, e contam com as seguintes colunas:

* **idade** = idade do cliente
* **sexo** = sexo do cliente (F ou M)
* **dependentes** = número de dependentes do cliente
* **escolaridade** = nível de escolaridade do clientes
* **salario_anual** = faixa salarial do cliente
* **tipo_cartao** = tipo de cartao do cliente
* **qtd_produtos** = quantidade de produtos comprados nos últimos 12 meses
* **iteracoes_12m** = quantidade de iterações/transacoes nos ultimos 12 meses
* **meses_inativo_12m** = quantidade de meses que o cliente ficou inativo
* **limite_credito** = limite de credito do cliente
* **valor_transacoes_12m** = valor das transações dos ultimos 12 meses
* **qtd_transacoes_12m** = quantidade de transacoes dos ultimos 12 meses

# 4 - Conclusão.

**Análise de Dados - Insights Extraídos:**



1. A maioria dos clientes possui renda de até 40k.

2. A escolaridade não parece influenciar nos limites de crédito e no tipo de cartão.

3. O tipo de cartão não impacta nos maiores limites, mas nos menores, predomina o cartão tipo blue.

4. Nos top 10 maiores limites de crédito, 9 pertencem a clientes do sexo masculino.

5. Nos top 10 menores limites de crédito, 8 pertencem a clientes do sexo feminino.

6. Não existem mulheres cadastradas no banco com salário anual superior a 60k.

7. A média de limite de crédito para mulheres com salários anuais inferiores a 40k é 20% menor que a média masculina.

8. Na média de limites com salários na faixa de '40k - 60k', os homens têm limites de crédito 25% maiores.
